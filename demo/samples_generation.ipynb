```python
import pickle as pk
import os
import numpy as np
```

```python
dataset_dir = "../data/Dataset_Lidar_20240111_vegas_ontrack_01/"
```

```python
# Reading the PKL file
with open(os.path.join(dataset_dir,"kittilidar_dbinfos_train.pkl"), 'rb') as file:
    data = pk.load(file)

# Now, 'data' contains the contents of the PKL file
print(data)
```

```python
import numpy as np
import open3d as o3d

def load_pointcloud(file_path):
    # Read the binary file
    pointcloud = np.fromfile(file_path, dtype=np.float32)
    
    # Each point is represented by 4 values (X, Y, Z, I)
    # Reshape the data to Nx4 where N is the number of points
    pointcloud = pointcloud.reshape(-1, 4)
    
    return pointcloud

def visualize_pointcloud(pointcloud):
    # Create an Open3D PointCloud object
    pcd = o3d.geometry.PointCloud()
    
    # Assign point cloud coordinates
    pcd.points = o3d.utility.Vector3dVector(pointcloud[:, :3])
    
    # Optionally assign colors or intensity values
    # For visualization purposes, we'll just set the color based on intensity
    intensity = pointcloud[:, 3]
    # Normalize intensity to range [0, 1] for visualization
    intensity = (intensity - intensity.min()) / (intensity.max() - intensity.min())
    # Convert intensity to RGB format
    colors = np.stack([intensity, intensity, intensity], axis=1)
    pcd.colors = o3d.utility.Vector3dVector(colors)
    
    # Visualize 
    o3d.visualization.draw_geometries([pcd])
```

```python
pointcloud = np.fromfile(os.path.join(dataset_dir,data["Car"][4]["path"]), dtype=np.float32)
    
# Each point is represented by 4 values (X, Y, Z, I)
# Reshape the data to Nx4 where N is the number of points
pointcloud = pointcloud.reshape(-1, 4)
visualize_pointcloud(load_pointcloud(os.path.join(dataset_dir,data["Car"][100]["path"])))
```

```python
tot = 0
for car in data["Car"]:
    if int(car["num_points_in_gt"])>150:
        visualize_pointcloud(load_pointcloud(os.path.join(dataset_dir,car["path"])))
        tot+=1
```
